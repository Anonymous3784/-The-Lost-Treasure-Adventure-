```python
The Lost Treasure Adventure - FUll game (Under 1 hour)

1. Define rooms & maps
- Used a dictionary

2. Initialize game state
-position/room?
-inventory
-score
-game_end?

3. Main game loop
-check for inventory
-show available direction..

4. validate movements
5. Enemy counter
6. Win & End conditions
```

```python
# The Lost Treasure Adventure - Colab Compatible Version

def play_game():
    # Define the rooms and map
    rooms = {
        "start": {
            "desc": "You are at the start of your adventure. Paths lead left and right.",
            "left": "forest",
            "right": "cave"
        },
        "forest": {
            "desc": "Tall trees surround you. You spot something shiny.",
            "right": "start",
            "item": "sword"
        },
        "cave": {
            "desc": "A dark cave. You hear a growl deeper inside. Caution!",
            "left": "start",
            "forward": "treasure_room"
        },
        "treasure_room": {
            "desc": "You found the hidden treasure! Congratulations!"
        }
    }

    # Initialize game state
    current_room = "start"
    inventory = []
    score = 0
    game_over = False

    # Ask player name
    name = input("Enter your name, adventurer: ")
    print(f"\nWelcome, {name}! Your adventure begins now....")

    # Main game loop
    while not game_over:
        print("\n" + "=" * 40)
        room = rooms[current_room]
        print(f"\nLocation: {current_room.upper()}")
        print(room["desc"])

        # Check for items
        if "item" in room and room["item"] not in inventory:
            print(f"\nYou found a {room['item']}!")
            take = input("Do you want to pick it up? (yes/no): ").lower()
            if take == "yes":
                inventory.append(room["item"])
                score += 10
                print(f"{room['item']} added to your inventory. Score +10!")

        # Show available directions
        directions = [k for k in room.keys() if k not in ["desc", "item"]]
        print("\nAvailable directions:", directions)
        move = input("Where do you want to go? ").lower()

        # Validate movement
        if move in room:
            current_room = room[move]
        else:
            print("You can't go that way. Try again.")
            continue

        # Enemy encounter
        if current_room == "cave":
            if "sword" not in inventory:
                print("\nA monster attacks! You have no weapon.")
                print("Game Over. üò¢")
                print(f"Final Score: {score}")
                game_over = True
                break
            else:
                print("\nA monster attacks!")
                print("You slay the monster with your sword! üó°Ô∏è Score +20")
                score += 20

        # Win condition
        if current_room == "treasure_room":
            print(f"\nüéâ Congratulations {name}, you found the treasure! üéâ")
            print(f"Your final score: {score}")
            game_over = True

            # Replay option
            play_again = input("\nDo you want to play again? (yes/no): ").lower()
            if play_again == "yes":
                play_game()
            else:
                print("Thanks for playing, adventurer!")
            break

# Run the game
play_game()
```
